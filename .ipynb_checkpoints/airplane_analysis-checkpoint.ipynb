# Data Cleaning

## Goal:
To prepare the dataset for analysis by ensuring consistency, handling missing data, and organizing the structure for efficient exploration.

---

## Steps:

1. **`.info()`**
   - View a summary of the DataFrame.
   - Check:
     - Number of non-null values
     - Data types
     - Column names
     - Shape (rows and columns)

2. **`.isna().sum()`**
   - Identify columns with missing values.
   - Count how many missing values exist per column.

3. **`.describe()`**
   - Generate summary statistics (mean, median, std, min, max, etc.) for numerical columns.

4. **`.sort_values()`**
   - Sort data in ascending or descending order.
   - Useful for identifying extremes (e.g., highest fatalities or destruction rates).

5. **Fixing Data Types**
   - Convert columns to appropriate types (e.g., dates to datetime, numeric strings to integers/floats).

6. **Standardizing Text**
   - Clean text columns (e.g., make all text lowercase, remove whitespace).
   - Useful for grouping and comparison.

7. **Checking for Duplicates**
   - Use `.duplicated()` and `.drop_duplicates()` to remove duplicate rows that may affect analysis accuracy.

---

Once the data is cleaned and structured, it will be ready for effective exploratory data analysis (EDA).

```python
#cleaning my data
# Check % of missing data
missing = aviation_df.isnull().mean().sort_values(ascending=False)
print("Columns with high missing data:\n", missing[missing > 0.5])
```

```python
#dropping the columns with high missing values
columns_to_be_dropped= ['Schedule' ,'Air.carrier', 'FAR.Description','Aircraft.Category']
aviation_df.drop(columns=columns_to_be_dropped, inplace=True)
columns_to_be_dropped
```

```python
#confirming that the columns above have been dropped
aviation_df.info()
```

```python
#fixing data types
aviation_df['Event.Date'] = pd.to_datetime(aviation_df['Event.Date'], errors='coerce')
aviation_df['Number.of.Engines'] = aviation_df['Number.of.Engines'].astype('Int64')
```

```python
#standardizing text
aviation_df['Make'] = aviation_df['Make'].str.strip().str.title()
aviation_df['Country'] = aviation_df['Country'].str.upper()
```

```python
#checking for duplicates
aviation_df.duplicated().sum()
```

Filling columns important for my analysis using .fillna()
my critical identifiers are make, model, total fatal injuries, minor injuries, uninjured, aircraft damage

```python
numeric_values=['Injury.Severity','Total.Fatal.Injuries','Total.Serious.Injuries','Total.Minor.Injuries','Total.Uninjured']
aviation_df[numeric_values]=aviation_df[numeric_values].fillna(0)
dummy_variables=['Injury.Severity','Aircraft.damage','Make','Model','Weather.Condition']
aviation_df[dummy_variables]= aviation_df[dummy_variables].fillna('Unknown')
```

Grouping and summarizing data. Focussing on the most relevant groupings for identifying low risk aircrafts for my client to invest in. The most relevant groupings include; the make , the model, the njury severity, total fatal injuries.

```python
#grouping by make to find which 'make' of aircraft is the safest.
make=aviation_df.groupby('Make').agg({
    'Event.Id': 'count',
    'Total.Fatal.Injuries': 'sum',
    'Aircraft.damage': lambda x: (x == 'Destroyed').sum()
})
#grouping by injury fatalities to find out which crashes ended up in fatalities
injury_group = aviation_df.groupby('Injury.Severity')['Event.Id'].count()
#grouping by weather condition to find out whether weather was a factor in causing these crashes''
weather_condition=aviation_df.groupby('Weather.Condition').agg({
    'Event.Id':'count',
    'Total.Fatal.Injuries': 'sum'})

```

# Visualizations Using Matplotlib

Visualizing aircraft crash data using **matplotlib** makes complex patterns easier to understand. It helps in **decision-making** and reveals important **trends** and **patterns** related to aviation safety.

---

## 1. Aircraft Makes Most Likely to Be Destroyed
This bar chart shows the **top 10 aircraft makes** with the highest destruction rate in crashes.  
- A **higher destruction rate** may indicate weaker structural integrity or higher-risk operating environments.

---

## 2. Aircraft Makes Least Likely to Be Destroyed
These aircraft makes have the **lowest destruction rates**, which likely means:
- They are more **durable**, or
- Involved in **fewer severe accidents**.

This visualization helps identify **potentially safer investment options**.

---

## 3. Aircraft with the Most Incidents
This bar chart highlights the aircraft makes with the **highest number of recorded incidents**.  
- A high number may suggest:
  - **Maintenance issues**
  - **Poor safety records**
  - Or simply **greater popularity** (more planes in service).

---

## 4. Aircraft with the Most Fatalities
Displays aircraft makes associated with the **highest total fatalities**.  
- This is critical for identifying aircraft with **poor survivability** in crashes.

---

## 5. Aircraft with the Least Fatalities
Shows aircraft makes with the **lowest non-zero fatality counts**.  
- Aircraft with **0 fatalities** were excluded.  
- Helps identify **safer aircraft models**.

---

## 6. Crashes Influenced by Weather Conditions
Analyzes the number of crashes where **weather played a role**.  
- Important for assessing **environmental risk factors**.

---

## 7. Number of Aircraft Incidents per Year
A time-series plot showing **how incidents have changed annually**.  
- Useful for spotting **trends or improvements in safety** over time.

---

## 8. Aircraft Incidents by Weather Condition Over the Years
Combines weather categories with time to show:
- **Weather-related incident trends**
- **Seasonal or climate-based risk factors**

---

## 9. Number of Fatalities per Incident
Calculates the **fatality rate per crash**.  
- Helps differentiate between:
  - High-fatality, low-frequency incidents  
  - Low-fatality, high-frequency incidents

---

## 10. Fatalities Over Time by Aircraft Make
A historical look at **fatalities over time**, grouped by aircraft make.  
- Shows which makes are **consistently safer or riskier** over the years.

---

**Summary:** These visualizations help uncover patterns, guide aviation safety policy, and inform aircraft investment or regulatory decisions.


```python
# Damage done on the aircraft. Which aircraft is more likely to be destroyed in a crash?
destroyed_aircrafts=aviation_df[aviation_df['Aircraft.damage'] == 'Destroyed'] #aircrafts that were destroyed
# Calculate destruction rate per aircraft make
destruction_rate = destroyed_aircrafts['Make'].value_counts() / aviation_df['Make'].value_counts()

# Drop missing values and sort
destruction_rate = destruction_rate.dropna().sort_values(ascending=False)

# Plot the top 10 makes by destruction rate
destruction_rate.head(10).plot(kind='bar', color='blue')
plt.title('Aircraft Makes most likely to be destroyes in a c')
plt.xlabel('Aircraft Make')
plt.ylabel('Destruction Rate')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#which aircraft by make is least likely to be destroyed in a crash
# Step 1: Count total incidents and destroyed incidents per make
total_incidents_per_make = aviation_df['Make'].value_counts()
destroyed_per_make = aviation_df[aviation_df['Aircraft.damage'] == 'Destroyed']['Make'].value_counts()

# Step 2: Calculate destruction rate
destruction_rate = destroyed_per_make / total_incidents_per_make
destruction_rate = destruction_rate.dropna()

# Step 3: Show aircraft makes with lowest destruction rates
lowest_destruction = destruction_rate.sort_values().head(10)

# Step 4: Plot
lowest_destruction.plot(kind='bar', color='green')
plt.title('Top 10 Aircraft Makes Least Likely to be Destroyed in a Crash')
plt.xlabel('Aircraft Make')
plt.ylabel('Destruction Rate')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Which aircraft has the highest number of incidents
aviation_df['Make'].value_counts().head(10).plot(kind='bar', color='red')
plt.title('Aircraft with highest number of incidents')
plt.xlabel('Aircraft Make')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=90)
plt.show()
#a high number of incidents is not ideal so according to this visualization Cessna and piper have very high incidents while mooney, robinsin ,grumman and the likes have low incidents which is ideal
```

```python
#which aircraft has the most fatalities?
total_fatalities=aviation_df.groupby('Make')['Total.Fatal.Injuries'].sum()
total_fatalities.dropna().sort_values(ascending=False)

#plot
total_fatalities.head(10).plot(kind='bar', color='orange')
plt.title(' Aircraft with the most fatalities')
plt.xlabel('Aircraft make')
plt.ylabel('Total Fatalities')
plt.xticks(rotation=90)
plt.show()
```

```python
#which aircraft has the least fatalities?
# Total fatalities per Make
total_fatalities = aviation_df.groupby('Make')['Total.Fatal.Injuries'].sum()

# Drop NaNs and filter to only real makes (not blank or unknown)
total_fatalities = total_fatalities.dropna()
total_fatalities = total_fatalities[total_fatalities.index.notnull()]
total_fatalities = total_fatalities[total_fatalities > 0] 

# Sort ascending for least fatal makes
least_fatalities = total_fatalities.sort_values().head(5)

# Plot
least_fatalities.plot(kind='bar', color='green')
plt.title('Aircraft Makes with the Least (Non-Zero) Total Fatalities')
plt.xlabel('Aircraft Make')
plt.ylabel('Total Fatalities')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
#Is weather condition a factor in causing the crashes?
aviation_df['Weather.Condition'].unique()
```

```python
crashes =aviation_df[aviation_df['Investigation.Type'] == 'Accident']
weather_crashes = crashes['Weather.Condition'].value_counts()
#plotting 
weather_crashes.plot(kind='bar', color='skyblue')
plt.title('Crashes influenced by weather condition')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=90)
plt.show()
```

```python
#which year had the most incidents?
aviation_df['Year'] = aviation_df['Event.Date'].dt.year
incidents_per_year = aviation_df['Year'].value_counts().sort_index()
most_incidents_count = incidents_per_year.max()
#plotting
incidents_per_year.plot(kind='line', color='skyblue', figsize=(10, 5))
plt.title('Number of Aircraft Incidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=90)
plt.show()
```

```python
#does weather condition have an impact on the number of incidents each year?
weather_by_year = aviation_df.groupby(['Year', 'Weather.Condition']).size().unstack()

# Plot
weather_by_year.plot(kind='line', figsize=(12, 6))
plt.title('Aircraft Incidents by Weather Condition Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.legend(title='Weather Condition')
plt.show()
```

```python
#fatalities over time with relation to aircraft make
fatalities_by_make_year = aviation_df.groupby(['Year', 'Make'])['Total.Fatal.Injuries'].sum().unstack()
aviation_df.plot(kind='line', figsize=(12, 6))
plt.title('Fatalities Over Time by Aircraft Make')
plt.xlabel('Year')
plt.ylabel('Total Fatal Injuries')
plt.grid(True)
plt.legend(title='Aircraft Make')
plt.show()
```

```python
#histogram of fatal injuries.
fatalities = aviation_df['Total.Fatal.Injuries'].dropna()

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(fatalities, bins=30, color='tomato', edgecolor='black')
plt.title('Distribution of Fatal Injuries in Aircraft Incidents')
plt.xlabel('Number of Fatalities per Incident')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Key Findings

- **Cessna** and **Piper** had the highest number of incidents, suggesting either high operational exposure or potential safety concerns.
- **Kenetic Aviation** and **Mcnicholas** had very few incidents and no recorded fatalities or destruction, making them strong candidates for low-risk investment.

- **Aircraft damage analysis** revealed that certain makes are more likely to be completely destroyed in crashes, potentially indicating weaker structural integrity or safety performance.
- **Destruction rate analysis** highlighted aircraft makes that consistently fared better in crash survivability, which is a valuable indicator of safer design.
